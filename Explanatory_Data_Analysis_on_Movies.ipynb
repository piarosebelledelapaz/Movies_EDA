1. __Import__ the movies dataset from the CSV file "movies_complete.csv". __Inspect__ the data.

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

# Fixing, cleaning and organizing the dataset

```python
from IPython.display import HTML
```

```python
movies_best = movies[["poster_path", "title", "budget_musd", "revenue_musd", 
                      "vote_count", "vote_average", "popularity"]].copy()
```

```python
movies_best
```

```python
# Add new column for profit, where profit = revenue-budget

movies_best["profit_musd"] = movies.revenue_musd.sub(movies.budget_musd)
```

```python
# Add new column for return of investment, where ROI = revenue/budget

movies_best["return"] = movies.revenue_musd.div(movies.budget_musd)
```

```python
movies_best
```

```python
# Rename columns

movies_best.columns=["", "Title", "Budget", "Revenue", "Votes", 
                    "Average_Rating", "Popularity", "Profit", "ROI"]
```

```python
# Set index to Title

movies_best.set_index("Title", inplace=True)
```

```python
# Load images from src path as HTML

subset= movies_best.iloc[:5,:2]
HTML(subset.to_html(escape=False))
```

```python
# Sort data by according to highest average rating (descending order)

movies_best.sort_values(by="Average_Rating", ascending=False)
```

```python
# Sort data by according to highest ROI (descending order)

movies_best.sort_values(by="ROI", ascending=False)
```

```python
# Filter all movies and get movies with budget >= 5 million and sort them by highest ROI

movies_best.loc[movies_best.Budget >= 5].sort_values(by="ROI", ascending=False)
```

```python
# Fill in NaN values

movies_best.Budget.fillna(0, inplace=True)
movies_best.Votes.fillna(0, inplace=True)
```

```python
movies_best.info()
```

## The best and the worst movies...

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

__Define__ an appropriate __user-defined function__ to reuse code.

```python
def best_worst_movies(n, by, ascending=False, min_bud=0, min_votes=0):
    
    new_movies = movies_best.loc[(movies_best.Budget >=min_bud)&(movies_best.Votes >= min_votes),
                                ["", by]].sort_values(by=by, ascending=ascending).head(n).copy()
    
    return HTML(new_movies.to_html(escape=False))
```

__Movies Top 5 - Highest Revenue__

```python
best_worst_movies(n=5, by="Revenue")
```

__Movies Top 5 - Highest Budget__

```python
best_worst_movies(n=5, by="Budget")
```

__Movies Top 5 - Highest Profit__

```python
best_worst_movies(n=5, by="Profit")
```

__Movies Top 5 - Lowest Profit__

```python
best_worst_movies(n=5, by="Profit", ascending=True)
```

__Movies Top 5 - Highest ROI__

```python
best_worst_movies(n=5, by="ROI", min_bud = 50)
```

__Movies Top 5 - Lowest ROI__

```python
best_worst_movies(n=5, by="ROI", ascending=True, min_bud=100)
```

__Movies Top 5 - Most Votes__

```python
best_worst_movies(n=5, by="Votes")
```

__Movies Top 5 - Highest Rating__

```python
best_worst_movies(n=5, by="Average_Rating", min_votes=100)
```

__Movies Top 5 - Lowest Rating__

```python
best_worst_movies(n=5, by="Average_Rating", ascending=True, min_bud=20, min_votes=100)
```

__Movies Top 5 - Most Popular__

```python
best_worst_movies(n=5, by="Popularity")
```

## Most Successful Franchises

5. __Find__ the __most successful Franchises__ in terms of

- __total number of movies__
- __total & mean budget__
- __total & mean revenue__
- __mean rating__

```python
movies.belongs_to_collection
```

```python
movies.belongs_to_collection.value_counts()
```

```python
franchises = movies.groupby("belongs_to_collection").agg({"title":"count", "budget_musd": ["sum", "mean"], 
                                                      "revenue_musd": ["sum", "mean"],
                                                      "vote_average": "mean", "popularity": "mean",
                                                      "ROI":"median", 
                                                      "vote_count":"mean"})
```

```python
franchises
```

```python
#total number of movies

franchises.nlargest(20, ("title", "count"))
```

```python
#total number of budget
franchises.nlargest(20, ("budget_musd", "mean"))
```

```python
#total number of revenue
franchises.nlargest(20, ("revenue_musd", "mean"))
```

```python
#total number of vote_count
franchises[franchises[("vote_count", "mean")] >=1000].nlargest(20, ("vote_average", "mean"))
```
